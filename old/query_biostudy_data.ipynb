```python
import requests
import yaml
```

```python
# Base URL for BioStudies API
base_url = 'https://www.ebi.ac.uk/biostudies/api/v1'

# Search for a specific study (replace study_id with the desired BioStudies ID)
study_id = 'E-MEXP-21'  # Example study ID
url = f'{base_url}/studies/{study_id}'

# Send the request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    study_data = response.json()
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
print(yaml.dump(study_data))
```

```python
for k, v in study_data['section'].items():
    print(f'{k}: {v}')
    print()
```
