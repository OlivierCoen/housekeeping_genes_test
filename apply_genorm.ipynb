```python
from genorm import m_measure, genorm
import pandas as pd
import numpy as np
```

```python
counts = pd.DataFrame(
    np.random.randint(1, 10, size=(20, 10)),
    columns=[f'gene_{s}' for s in 'abcdefghij'],
    index=[f'sample_{i}' for i in range(20)],
)
```

```python
counts
```

```python
# Compute `M` value for this set of control genes.
m_measure(counts)
```

```python

# Select top 2 control genes with lowest `M`.
gene_names, m_values = genorm(counts, n_stop=5)
```
